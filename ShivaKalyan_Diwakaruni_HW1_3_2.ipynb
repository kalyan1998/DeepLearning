```python
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
train_dataset = datasets.MNIST('./data', train=True, download=True, transform=transform)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_dataset = datasets.MNIST('./data', train=False, download=True, transform=transform)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=True)
```

```python
def train_model(model, optimizer):
    model.train()
    train_losses, test_losses = [], []
    train_accuracies, test_accuracies = [], []
    epochs = 10
    for epoch in range(epochs):
        running_loss = 0.0
        correct_train = 0
        total_train = 0
        for i, data in enumerate(train_loader, 0):
            inputs, labels = data
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total_train += labels.size(0)
            correct_train += (predicted == labels).sum().item()
        
        model.eval()
        test_loss = 0
        correct_test = 0
        total_test = 0
        with torch.no_grad():
            for data, target in test_loader:
                outputs = model(data)
                test_loss += criterion(outputs, target).item()
                _, predicted = torch.max(outputs.data, 1)
                total_test += target.size(0)
                correct_test += (predicted == target).sum().item()
        
        print(f'Epoch {epoch+1}, Train Loss: {running_loss/len(train_loader):.4f}, Test Loss: {test_loss/len(test_loader):.4f}')
        test_losses = (test_loss / len(test_loader))
        test_accuracy = 100 * correct_test / total_test
        
        train_losses = (running_loss / len(train_loader))
        train_accuracy = 100 * correct_train / total_train
    
    return train_losses, test_losses, train_accuracy, test_accuracy, sum(p.numel() for p in model.parameters())
```

```python
def no_params(model):
    return sum(p.numel() for p in model.parameters())
def define_optimizer(model):
    return optim.Adam(model.parameters(), lr = 0.0001)
```

```python
model_classes = [Net_1, Net_2, Net_3, Net_4, Net_5, Net_6, Net_7, Net_8, Net_9, Net_10, Net_11]
model_names = ['Model 1', 'Model 2', 'Model 3', 'Model 4', 'Model 5', 'Model 6', 'Model 7', 'Model 8', 'Model 9', 'Model 10', 'Model 11']
final_train_losses, final_test_losses, final_train_accuracies, final_test_accuracies, final_param_arr = [],[],[],[], []
for model_class, model_name in zip(model_classes, model_names):
    model = model_class()
    criterion = nn.CrossEntropyLoss()
    torch.manual_seed(1)
    print(f"{model_name} parameters: ", no_params(model))
    optimizer = define_optimizer(model)
    train_losses, test_losses, train_accuracies, test_accuracies, param_arr = train_model(model, optimizer)
    final_train_losses.append(train_losses)
    final_test_losses.append(test_losses)
    final_train_accuracies.append(train_accuracies)
    final_test_accuracies.append(test_accuracies)
    final_param_arr.append(param_arr)

```

```python
# Plotting
plt.figure(figsize=(12, 5))
plt.scatter(final_param_arr, final_train_losses,color = "r")
plt.scatter(final_param_arr, final_test_losses, color = "b")
plt.xlabel('Parameters')
plt.ylabel('Loss')
plt.legend(['TRAINING_LOSS', 'TEST_LOSS'])
plt.show()
```
